```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')

# Load the breast cancer dataset
df = pd.read_csv('archive/Breast_cancer_dataset.csv')
```

```python
print("First 5 rows:")
display(df.head())

print("\nBasic Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe())
```

```python
# 3. Check for Missing Values
print("\nMissing Values Count:")
display(df.isnull().sum())

# 4. Handle Missing Values Example
df_filled = df.fillna(df.mean(numeric_only=True))
```

```python
# 5. Check Duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Remove duplicates if any
df = df.drop_duplicates()
```

```python
# 6. Rename Columns (example)
df.rename(columns=lambda x: x.strip().lower().replace(' ', '_'), inplace=True)
```

```python
# 7. Data Selection
# Select specific columns
if len(df.columns) >= 3:
    display(df[df.columns[:3]].head())

# Filter rows (example: if dataset has 'diagnosis' column)
if 'diagnosis' in df.columns:
    malignant = df[df['diagnosis'] == 'M']
    benign = df[df['diagnosis'] == 'B']
    print(f"\nMalignant cases: {len(malignant)}, Benign cases: {len(benign)}")
```

```python
# 8. Sorting
if df.select_dtypes(include='number').shape[1] > 0:
    first_num_col = df.select_dtypes(include='number').columns[0]
    df_sorted = df.sort_values(by=first_num_col, ascending=False)
    print(f"\nSorted by {first_num_col}:")
    display(df_sorted.head())
```

```python
# 9. Grouping & Aggregation
if 'diagnosis' in df.columns:
    group_stats = df.groupby('diagnosis').mean(numeric_only=True)
    print("\nMean values grouped by diagnosis:")
    display(group_stats)

# 10. Value Counts
if 'diagnosis' in df.columns:
    print("\nDiagnosis counts:")
    display(df['diagnosis'].value_counts())
```

```python
# 11. Correlation Analysis
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 12. Basic Visualizations
# Histogram of first numeric column
if first_num_col:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[first_num_col], kde=True)
    plt.title(f"Distribution of {first_num_col}")
    plt.show()
```

```python
# Boxplot example
if 'diagnosis' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='diagnosis', y=first_num_col, data=df)
    plt.title(f"{first_num_col} by Diagnosis")
    plt.show()

# Pairplot for first few numeric columns
num_cols = df.select_dtypes(include='number').columns[:4]
if len(num_cols) > 1:
    sns.pairplot(df, vars=num_cols, hue='diagnosis' if 'diagnosis' in df.columns else None)
    plt.show()
```
